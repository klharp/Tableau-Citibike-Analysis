```python
# Dependencies
import pandas as pd
import os
import pprint
from pathlib import Path
from datetime import datetime
```

## Combine CSVs from a directory

```python
# Create variable for files in directory 
files = [f for f in os.listdir("../Tableau-HW-support/datafiles2/") if f.endswith(".csv")]

files
```

## Create dataframe for 2020 data

```python
# https://stackoverflow.com/questions/63886787/how-to-create-a-dataframe-from-multiple-csv-files
# Initialize and empty dataframe
all_data = pd.DataFrame()

# Iterate through files and contents, then concatenate the data into the dataframe
for file in files:
    df = pd.read_csv("../Tableau-HW-support/datafiles2/" + file)
    all_data = pd.concat([all_data, df])
    
all_data
# 4286872   rows, 15 columns
```

```python
#Remove any null data
citibike = all_data.dropna()

citibike
```

```python
#Drop rows were gender is unknown
citibike = citibike[citibike.gender != 0]

citibike
```

```python
# Replace gender 1 to male, 2 to female

citibike["gender"].replace({1:"male", 2:"female"}, inplace=True)

citibike.head(100)
```

```python
#Split date/time column to date & time
citibike["date_start"] = pd.to_datetime(citibike["starttime"]).dt.date
citibike["date_end"] = pd.to_datetime(citibike["stoptime"]).dt.date
citibike["time_start"] = pd.to_datetime(citibike["starttime"]).dt.time
citibike["time_end"] = pd.to_datetime(citibike["stoptime"]).dt.time

citibike
```

```python
#Export as CSV
citibike.to_csv("tableau/data/citibike2.csv", index=False, header=True)
```
